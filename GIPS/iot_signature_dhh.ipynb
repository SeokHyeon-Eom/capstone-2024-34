```python
import pickle
```

```python
with open('./res/iot_dhh_signature.pkl', 'rb') as f:
    signature = pickle.load(f)
    
with open('./res/iot_dhh_stopword.pkl', 'rb') as f:
    stopword = pickle.load(f)
```

```python
print('SIGNAURE')

for i in signature.values():
    for j in i:
        print(j, end=" ")
    print()
    
print('STOPWORD')
for i in stopword:
    print(i)
```

```python
filters = [i[0] for i in stopword if i[1] > 99]
```

```python
for cluster, sig in signature.items():
    for token in sig:
        if token[1] not in filters:
            print(token)
```
